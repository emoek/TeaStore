```python
import pandas as pd
import matplotlib.pyplot as plt
# Read the CSV file into a DataFrame
df = pd.read_csv("../results/resultv24/exp11_16_8_generic_powerScalability_1.csv")
df1 = pd.read_csv("../results/resultv24/exp11_16_8_generic_powerScalability_2.csv")
df2 = pd.read_csv("../results/resultv24/exp11_16_8_generic_powerScalability_3.csv")

# Drop the first 120 rows
df = df.iloc[120:]
df1 = df1.iloc[120:]
df2 = df2.iloc[120:]

# Calculate the mean of the "value" column
mean_value = df['value'].mean()
mean_value1 = df1['value'].mean()
mean_value2 = df2['value'].mean()

means = [mean_value, mean_value1, mean_value2]
print(means)
final = sorted(means)[1]


print("Mean value (excluding first 120 rows):", final)
```
