```python
import pandas as pd
import matplotlib.pyplot as plt
# Read the CSV file into a DataFrame
df = pd.read_csv("../results/resultv17/exp30_500_8_generic_powerScalability_1.csv")

# Drop the first 120 rows
df = df.iloc[120:]

# Calculate the mean of the "value" column
mean_value = df['value'].mean()

print("Mean value (excluding first 120 rows):", mean_value)
```

```python


# Data from the table
data = {
    'Load': [10, 10, 10, 10, 50, 50, 50, 50, 100, 100, 100, 100, 
             150, 150, 150, 150, 200, 200, 200, 200, 250, 250, 250, 250, 
             500, 500, 500, 500],
    'Resource': [1, 2, 4, 8]*7,
    'Mean Power Consumption': [292.49, 276.34, 288.68, 143.67, 
                                321.31, 354.60, 419.57, 388.92, 
                                320.95, 350.87, 425.63, 390.11, 
                                296.86, 384.85, 387.11, 417.65, 
                                271.05, 298.97, 354.34, 397.25, 
                                363.05, 354.33, 401.18, 371.81, 
                                292.86, 415.41, 419.04, 409.74]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plot
plt.figure(figsize=(10, 6))
for resource in df['Resource'].unique():
    plt.plot(df[df['Resource']==resource]['Load'], 
             df[df['Resource']==resource]['Mean Power Consumption'], 
             marker='o', label=f'Resource {resource}')

plt.title('Mean Power Consumption for Different Loads and Resources')
plt.xlabel('Load')
plt.ylabel('Mean Power Consumption')
plt.legend()
plt.grid(True)
plt.show()
```
